```python
def readscore(file):
    
    refID = 0 # id of reference sequence in that cluster
    ref = 0 # reference sequence in that cluster
    seqID = {} # id of sequences in the hmmsearchfile
    bias_full = {} # full - bias  of sequences alignment
    escore_full = {} # full - evalue of sequences alignment
    score_full = {} # full - score of sequences alignment
    
    bias_best = {} # best - bias  of sequences alignment
    escore_best = {} # best - evalue of sequences alignment
    score_best = {} # best - score of sequences alignment
    
    cnt = 0
    
    with open(file) as fp:
        for line in fp:
            if [char for char in line][0] != '#':
                
                linevector = line.split()

                seqID[cnt] = str(linevector[0].split("|")[1])

                escore_full[cnt] = float(linevector[4])
                score_full[cnt] = float(linevector[5])
                bias_full[cnt] = float(linevector[6])

                escore_best[cnt] = float(linevector[7])
                score_best[cnt] = float(linevector[8])
                bias_best[cnt] = float(linevector[9])
                
                cnt+=1
    
    # save data in structure
    data = {'sequences ID':seqID,'e-value full':escore_full,'score-full':score_full,'bias-full': bias_full,'simplified intervals': {},
            'e-value best':escore_best,'score-best': score_best,'bias-best': bias_best,'number of sequences':cnt,
            'tuples':{},'reference':{},'sequences per cluster': 0,'safety windows intervals':{},'safety windows': {},
           'safety positions': {},'interval cliques': {},'sequences': {},'average length':{},
           'coverage average': {},'length of cliques per tuples': 0,'number of cliques per tuples':0,'coverage of cliques per tuples':0,
            'overlaps': 0, 'overlapping length': [0]*cnt,'noise of cliques per tuples':0,'ratio lenght': 0,
            'taxonomy id':0,'sequences windows intervals': {},'cliques per tuples': {}, 'cliqueNonMaxPerTuple': {},
            'cliqueNonMaxTotal': 0,'nonMaxAverageLengthPerTuple': {}, 'nonMaxAverageCoveragePerTuple':{},
            'nonMaxAverageNoisePerTuple': {},'nonMaxAverageLength': 0,'nonMaxAverageCoverage':0,
            'nonMaxAverageNoise': 0,
           }
    
    
    return data;
```
