```python
from random import randint
def printCodedWindows(data,mypath,subcluster):
    
    test = open(mypath+'/ColoredWindows/'+subcluster+"_coded.txt",'w')
    
    # gather all safety windows
    safetyIntervals = data['safety windows intervals']
    nS = data['number of sequences']
    reference = data['reference']
    
    # set of Intervals
    setOfIntervals = set()
    referenceAsList = [char for char in reference]
    referenceAsColours = [0]*len(referenceAsList)
    
    for i in range(1,nS):
        listOfIntervals = safetyIntervals[i]
        
        for k in range(0,len(listOfIntervals)):
            setOfIntervals.add(listOfIntervals[k])
    
    associatedList =list(setOfIntervals)
    
    # get coded windows
    colorForWindows = []
    for i in range(0,len(associatedList)):
        colorForWindows.append('#%06X' % randint(0, 0xFFFFFF))
    
    print(reference,file=test)
    
    # print coded windows
    for i in range(1,nS):
        listOfIntervals = safetyIntervals[i]
        referenceAsColours = [0]*len(referenceAsList)
        for k in range(0,len(listOfIntervals)):
            safetyPair = listOfIntervals[k]
            getPositionForColor = associatedList.index(safetyPair)
            for l in range(int(safetyPair[0]),int(safetyPair[1])-1):
                referenceAsColours[l] = getPositionForColor
        for k in range(0,len(referenceAsList)):
            if referenceAsColours[k] == 0:
                print(referenceAsList[k],end="",file=test)
            else:
                print(color(referenceAsList[k],back = colorForWindows[referenceAsColours[k]]),end="",file=test)
        print("",file=test)
    
    return data;
```

```python
def scoredisplay(data,path,subcluster):
    
    # display graphic with same scores from MSA, phmmer and hmmsearch in the group
    
    import matplotlib.pyplot as plt
    
    nS = data['number of sequences']
    sequences = data['sequences ID']
    escore_full = data['e-value full']
    score_full = data['score-full']
    bias_full = data['bias-full']
    escore_best = data['e-value best']
    score_best = data['score-best']
    bias_best = data['bias-best']

    
    fig, axs = plt.subplots(3,figsize=(15,15))
    
    axs[0].plot(sequences.values(),escore_full.values(),'go')
    axs[0].set_title('E-value - Full Sequence')
    axs[0].tick_params(labelrotation=90)
    axs[1].plot(sequences.values(),score_full.values(),'ro')
    axs[1].set_title('Score - Full Sequence')
    axs[1].tick_params(labelrotation=90)
    axs[2].plot(sequences.values(),bias_full.values(),'bo')
    axs[2].set_title('Bias - Full Sequence')
    axs[2].tick_params(labelrotation=90)
    
    plt.subplots_adjust(wspace = 0.4, 
                    hspace=0.4)
    
    fig.savefig(path+'/Escore/'+'Safety_Windows_Full_Sequence'+subcluster+'.png')
    
    fig, axs = plt.subplots(3,figsize=(15,15))
    
    axs[0].plot(sequences.values(),escore_best.values(),'co')
    axs[0].set_title('E-value - Best 1 Domain')
    axs[0].tick_params(labelrotation=90)
    axs[1].plot(sequences.values(),score_best.values(),'mo')
    axs[1].set_title('Score - Best 1 Domain')
    axs[1].tick_params(labelrotation=90)
    axs[2].plot(sequences.values(),bias_best.values(),'yo')
    axs[2].set_title('Bias - Best 1 Domain')
    axs[2].tick_params(labelrotation=90)
    
    plt.subplots_adjust(wspace = 0.4, 
                    hspace=0.4)
    
    fig.savefig(path+'/Escore/'+'Safety_Windows_Full_Sequence'+subcluster+'.png')
    
    return data
```

```python
from colr import color
def printColouredWindows(data,mypath,subcluster):
    
    test = open(mypath+'/ColoredWindows/'+subcluster+"_colored.txt",'w')
    

    # print the windows in a different colour
    safetyIntervals = data['safety windows intervals'] # safety windows
    nS = data['number of sequences'] # number of sequences
    reference = data['reference'] # reference in that cluser
    
    referenceAsList = [char for char in reference]
    referenceAsColours = [0]*len(referenceAsList)
    print(reference,file=test)
    
    for i in range(1,nS):
        listOfIntervals = safetyIntervals[i]
        referenceAsColours = [0]*len(referenceAsList)
        for k in range(0,len(listOfIntervals)):
            safetyPair = listOfIntervals[k]
            for l in range(int(safetyPair[0]),int(safetyPair[1])-1):
                referenceAsColours[l] = 1
        for k in range(0,len(referenceAsList)):
            if referenceAsColours[k] == 0:
                print(referenceAsList[k],end="",file=test)
            else:
                print(color(referenceAsList[k],back='red'),end="",file=test)
        print("",file=test)
    
    return data;

```

```python
#cluster_name #length_of_reference
#number_of_sequences #average_length_sequences 
#number_of_safety_windows #average_length_safety_windows #average_coverage #average_noise
#number_of_cliques #average_clique_size #length_of_common_safety_window #coverage #noise
#number_of_tuples #optimal_size_of_tuples #length #coverage #noise

def compilationTable(data,subcluster,output):
    nS = data['sequences per cluster']
    R = len(data['reference'])
    setOfSequences = list(data['protein sequence'])
    safetyWindows = data['safety windows']
    lengthTuplesCliques = data['length of cliques per tuples'] 
    numberTuplesCliques = data['number of cliques per tuples'] 
    noiseTuplesCliques = data['noise of cliques per tuples']
    coverageTuplesCliques = data['coverage of cliques per tuples']
    
    averageLengthOfSafetyWindows = sum([len(safetyWindows[seq][0]) for seq in safetyWindows])/len(safetyWindows)
    averageCoverageOfSafetyWindows = averageLengthOfSafetyWindows/R
    averageLengthSequences = sum([len(seq) for seq in setOfSequences])/len(setOfSequences)

    print(subcluster," & ",R," & ",nS," & ", averageLengthSequences, " & ",
          len(safetyWindows)," & ", averageLengthOfSafetyWindows," & ", averageCoverageOfSafetyWindows," & ",1-averageCoverageOfSafetyWindows," & ",
          len(lengthTuplesCliques)," & ",
          sum(lengthTuplesCliques)/len(lengthTuplesCliques), " & ",
          sum(coverageTuplesCliques)/len(coverageTuplesCliques)," & ",
          sum(numberTuplesCliques)/len(numberTuplesCliques), "\\", 
          file = output,flush=True
         )
    
    
    return 0;
```
